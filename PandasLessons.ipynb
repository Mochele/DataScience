### FILTERING THE DATA

```python
pokemon.head(5)
```

```python
##Filtering based on 2 conditions

Fire_list=pokemon["Type 1"]=="Fire"
Grass_list=pokemon["Type 2"]=="Normal"
pokemon.loc[Fire_list & Grass_list]
```

```python
##what if i want to exclude observations with the name mega in them
#pokemon["Name"].str.contains("Mega")


pokemon.loc[~pokemon["Name"].str.contains("Mega")]
```

```python
## How can we get pokemons with type 1 Grass and Fire

import re

pokemon.loc[pokemon["Type 1"].str.contains("Fire|Grass", flags=re.I , regex=True)]
```

```python
## Getting pokemons with names starting with pi

pokemon.loc[pokemon["Name"].str.contains("^pi[a-z]*", flags=re.I , regex=True)]
```

### CONDITIONAL CHANGES

```python
# How can we conditionally change the names of observations within variables
##subset1=pokemon["Type 1"]=="Fire"
##pokemon.loc[subset1]="Flamer"

pokemon.loc[pokemon["Type 1"]=="Fire","Type 1"]="Flamer"

pokemon
```

```python
## Changing one column conditional on another column. e.g, If i decide that legendary pokemons are all pokemons of type fire

pokemon.loc[pokemon["Type 1"]=="Flamer","Legendary"]=True
pokemon
```

```python
## Changing multiple parameters at the same time

pokemon.loc[pokemon["Total"]>100,["Generation","Legendary"]]="Test"
pokemon
```

```python
## Changing multiple parameters at the same time: BUT CREATING INDIVIDUAL FILL IN'S FOR EACH VARIABLE SPECIFIED

pokemon.loc[pokemon["Total"]>100,["Generation","Legendary"]]=["Test 1", "Test 2"]
pokemon
```

### AGGREGATED STATISITCS-groupby

```python
##It can calculate the mean, sums, count

pokemon.groupby("Type 1").mean().sort_values("Defense", ascending=False)

# for the count function one can create a new variable containing the value 1, then do the grouped counting
#Can use multiple groups: pokemon.groupby(["Type 1","Type 2"]).().sort_values("Defense", ascending=False)
```

```python
## Creating a variable that counts:
pokemon["counter"]=1
pokemon.groupby(["Type 1"]).count()["counter"]
```

```python
# Continued grouping with count
pokemon.groupby(["Type 1", "Type 2"]).count()["counter"]
```
